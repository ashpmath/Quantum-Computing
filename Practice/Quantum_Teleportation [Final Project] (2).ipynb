```python
#First parameter represents the num of qubits, second represents the num of classical registers
circuit = QuantumCircuit(3, 3)
```

```python
#%matplotlib inline
circuit.draw(output = 'mpl')
```

```python
circuit.reset(0)
circuit.x(0)
```

```python
circuit.barrier()
```

```python
circuit.draw(output='mpl')
```

```python
#Apply Hadamard gate to qubit 1
circuit.h(1)

#Introduce CNOT gate to qubit 2
circuit.cx(1, 2)

circuit.draw(output = 'mpl')

#applies bell state
```

```python
#apply CNOT gate operator (control is q0, target qubit is q1) 
circuit.cx(0,1)

#apply Hadamard operator to q0. 
circuit.h(0) 

#draw the circuit
circuit.draw(output='mpl')
```

```python
#apply barrier
circuit.barrier()
```

```python
#apply measurement operators on q0 and q1
circuit.measure([0,1], [0,1])

#apply barrier
circuit.barrier()

#draw circuit
circuit.draw(output='mpl')
```

```python
#Apply controlled not gate to qubit 1 (control) and qubit 2 (target)
circuit.cx(1,2)

#appy Z gate
circuit.cz(0,2)

#draw circuit
circuit.draw(output='mpl')
```

```python
#measure results of qubit 2 and put into a classical register - bobs result
circuit.measure(2,2)

#Simulate circuit using one of the quantum simulators in the qasm simulator backend
simulator = Aer.get_backend('qasm_simulator')

# specify the number of shots - because the measurement of a qubit in a superposition 
# the state is random — so the outcome is sometimes 0 and sometimes 1 — must repeat the measurement 
# multiple times to determine the likelihood that a qubit is in a particular state
result = execute(circuit, backend=simulator, shots=1024).result()

#get the number of counts
counts = result.get_counts()

#import the function plot histogram
from qiskit.tools.visualization import plot_histogram, plot_state_qsphere, plot_bloch_multivector

#plot num of counts
plot_histogram(counts)
```

```python
#plot state qsphere - global visualization of all quantum states
circuit.save_statevector()
result = simulator.run(assemble(circuit)).result()
out_state = result.get_statevector()
print(out_state) # Display the output state vector

#plot_state_qsphere(circuit)

#plot bloch sphere - local view of the quantum state
plot_bloch_multivector(out_state)
```
